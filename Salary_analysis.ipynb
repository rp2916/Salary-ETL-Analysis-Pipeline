```python
import pandas as pd
import numpy as np

df = pd.read_csv('Salary_survey.csv')
print(df.head())
```

```python
print(df.info())
```

```python
name_map = {
    'How old are you?':'Age',
    'What industry do you work in?':'Industry',
    'Job title':'Job_title',
    'If your job title needs additional context, please clarify here:':'Job_Title_info',
    "What is your annual salary? (You'll indicate the currency in a later question. If you are part-time or hourly, please enter an annualized equivalent -- what you would earn if you worked the job 40 hours a week, 52 weeks a year.)":'Annual_salary',
    'How much additional monetary compensation do you get, if any (for example, bonuses or overtime in an average year)? Please only include monetary compensation here, not the value of benefits.':'Additional_compensation',
    'Please indicate the currency':'Currency',
    'If "Other," please indicate the currency here: ' : 'Other_currency',
    'If your income needs additional context, please provide it here:' : 'Income_info',
    'What country do you work in?' : 'Country',
    "If you're in the U.S., what state do you work in?" : 'Work_State_US',
    'What city do you work in?' : 'Work_City',
    'How many years of professional work experience do you have overall?' : 'Overall_Experience',
    'How many years of professional work experience do you have in your field?' : 'Experience_in_field',
    'What is your highest level of education completed?':'Highest_education',
    'What is your gender?': 'Gender',
    'What is your race? (Choose all that apply.)':'Race'
}

df = df.rename(columns=name_map)
print(df.columns)
```

```python
print(df.info())
```

```python
print(df.dtypes)
```

```python
df['Annual_salary'] = df['Annual_salary'].astype(str).str.replace(',','')
df['Annual_salary'] = pd.to_numeric(df['Annual_salary'],errors = 'coerce')
```

```python
df['Annual_salary'].head()
```

```python
df.info()
```

```python
df['Country'] = df['Country'].astype(str).str.lower().str.strip()
print(df['Country'].value_counts().head(20))
```

```python
country_map = {
    'united states' : 'usa',
    'us' : 'usa',
    'u.s.' : 'usa',
    'united states of america' : 'usa',
    'u.s.a' : 'usa',
    'united kingdom' : 'uk',
    'england' : 'uk',
    'scotland' : 'uk',
    'the netherlands':'netherlands',
    'u.s' : 'usa',
    'u.s.a.' : 'usa',
    'united state' : 'usa',
    'unites states' : 'usa',
    'america': 'usa',
    'nl' : 'netherlands',
    'nederland' :'netherlands',
    'canadw' : 'canada',
    'englang' : 'uk',
    'united statew' : 'usa',
    'u.k. (northern england)' : 'uk',
    'u.k' : 'uk',
    'u.k.' : 'uk',
    'united stated' : 'usa'
}

df['Country'] = df['Country'].replace(country_map)
```

```python
print(df['Country'].value_counts().tail(30))
```

```python
df.loc[df['Country'].str.contains('states|usa|america|u.s', na = False), 'Country'] = 'usa'
df.loc[df['Country'].str.contains('kingdom|uk|gb|england|wales|scotland', na = False), 'Country'] = 'uk'
df.loc[df['Country'].str.contains('canada|canad', na = False), 'Country'] = 'canada'
df.loc[df['Country'].str.contains('uae|emirates', na = False), 'Country'] = 'uae'
```

```python
print(df['Country'].value_counts().tail(30))
```

```python
df.loc[df['Country'].str.contains('stat|stt|u. s|u s|uniter', na=False), 'Country'] = 'usa'
df.loc[df['Country'] == 'virginia', 'Country'] = 'usa'
df.loc[(df['Country'].str.len() > 20) & (df['Country'] != 'usa'), 'Country'] = 'other'
df['Country'] = df['Country'].str.lower().str.strip()
```

```python
print(df['Country'].value_counts().tail(30))
```

```python
final_fixes = {
    'us of a': 'usa',
    'csnada': 'canada',
    'can': 'canada',
    'uxz': 'usa',
    'hartford': 'usa',
    'new zealand aotearoa': 'new zealand',
    'u.a.': 'other',
    'international': 'other',
    'brasil': 'brazil',
    'panamá': 'panama'
}
df['Country'] = df['Country'].replace(final_fixes)
df.loc[df['Country'].str.contains('philippines', na=False), 'Country'] = 'philippines'
```

```python
print(df['Country'].value_counts().tail(30))
```

```python
country_counts = df['Country'].value_counts()
major_countries = country_counts[country_counts >= 10].index
outliers = country_counts[country_counts < 10]
print(f"Number of Major Countries: {len(major_countries)}")
print(f"Number of Unique Outliers: {len(outliers)}")
print(f"Total Rows considered 'Outliers': {outliers.sum()}")
print(outliers.head(20))
```

```python
df['Exp_Years_Clean'] = df['Overall_Experience'].astype(str).str.extract(r'(\d+)').astype(float)
df['Field_Exp_Clean'] = df['Experience_in_field'].astype(str).str.extract(r'(\d+)').astype(float)
print(df[['Overall_Experience','Experience_in_field']].head(10))
```

```python
print(df['Currency'].value_counts())
```

```python
df_usd = df[df['Currency'] == 'USD'].copy()
```

```python
df_usd = df_usd.dropna(subset=['Annual_salary'])
```

```python
print(df_usd.info())
```

```python
final_columns = [
    'Age', 'Industry', 'Job_title', 'Annual_salary', 
    'Additional_compensation', 'Country', 'Work_State_US', 'Work_City',
    'Highest_education', 'Gender', 'Race', 'Exp_Years_Clean', 'Field_Exp_Clean'
]

df_final = df_usd[final_columns].copy()
df_final['Additional_compensation'] = df_final['Additional_compensation'].fillna(0)
```

```python
from sqlalchemy import create_engine
USER = 'postgres'
PASSWORD = 'Samu2916'
HOST = 'localhost'
PORT = '5432'
DB_NAME = 'Salary_analysis'

# 2. Create the engine
connection_string = f'postgresql://{USER}:{PASSWORD}@{HOST}:{PORT}/{DB_NAME}'
engine = create_engine(connection_string)

# 3. Send the DataFrame to Postgres
# name='salaries' is what the table will be called in SQL
# if_exists='replace' means it will overwrite the table if you run it twice
df_usd.to_sql(name='salaries', con=engine, if_exists='replace', index=False)

print("Data successfully loaded into PostgreSQL!")
```
